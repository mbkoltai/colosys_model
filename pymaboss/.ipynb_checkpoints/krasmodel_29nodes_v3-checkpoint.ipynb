```python
import maboss
import ginsim
import pypint

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model = maboss.load("charitemodel_v5.bnd", "charitemodel_v2_cdk1_relaxed.cfg")
```

```python
# set cell fate+other nodes to 0
nodes=('Senescence','mitotic_catastrophe','Mitosis','Apoptosis_CC3',
  'CDC25A','CDC25B','CDC25C','CDK1','CDK2','CDK4_6',
       'MK2_HSP27','K_N_HRAS','RPS6KA1_CSNK2A1','MEK','ERK1_2','p38',
       'ATR','ATM','TP53','CHEK1','CHEK2','HRR2','FAHRR')
istate=[1,0]
for nd in nodes:
    model.network.set_istate(nd, istate)

# set input nodes to 1
model.network.set_istate('GFR', [0,1])
model.network.set_istate('DSB_SSB', [0,1])

# model.get_logical_rules()
```

```python
# run simul
maboss.set_output(model,['Apoptosis_CC3','Senescence','mitotic_catastrophe','Mitosis','mitotic_catastrophe',
    'HRR2','DSB_SSB','MK2_HSP27','TP53','ATM','ATR','CHEK1','CHEK2','p38','ERK1_2','BRAF','K_N_HRAS',
    'CDC25A','CDC25B','CDC25C','CDK1','CDK2','CDK4_6'])
model.update_parameters(discrete_time=0, use_physrandgen=0,
                            seed_pseudorandom=100, sample_count=500,
                            max_time=30, time_tick=0.1, thread_count=4,
                            statdist_traj_count=100, statdist_cluster_threshold=0.9)
master_results = model.run()
```

```python
nodes_probtraj=master_results.get_nodes_probtraj()
nodes_probtraj_df=pd.DataFrame(master_results.get_nodes_probtraj())
# plots of cell cycle regulators
col_inds=nodes_probtraj_df.columns.str.contains('|'.join(['CDK','CDC25']),case=False)
nodes_probtraj_df.loc[:,col_inds].plot(linewidth=3) 
fig_nodetraj=plt.gcf(); fig_nodetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```

```python
# plot of cell fate nodes
sel_vars=['RAS','BRAF','CHEK1','MK2','mitos','apopt','senesc','DSB']
# ['mit','apopt','senesc','DSB','CDK1','TP53','CHEK','MK2','ATM','ATR','AKT']
nodes_probtraj_df.loc[:,nodes_probtraj_df.columns.str.contains('|'.join(sel_vars),case=False)].plot(linewidth=3) 
fig_nodetraj=plt.gcf(); fig_nodetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5),loc="center left",borderaxespad=0)
```

```python
# show states that have >0.1 prob at any time point
states_probtraj=master_results.get_states_probtraj()
states_probtraj.loc[:,(np.sum(states_probtraj>0.05)>0) & (states_probtraj.columns!='<nil>')].plot(linewidth=3)
fig_statetraj=plt.gcf(); fig_statetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5),loc="center left",borderaxespad=0)
```

```python
# show trajectory of states that have positive probability 
col_inds=(states_probtraj.iloc[-1,:]>0.01) & (states_probtraj.columns!='<nil>') & \
    states_probtraj.columns.str.contains('mit',case=False)
states_probtraj.loc[:,col_inds].plot(linewidth=3)
fig_statetraj=plt.gcf(); fig_statetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5),loc="center left",borderaxespad=0)
```

```python
# piechart of fixpoints
master_results.plot_piechart()
fig_state_piechart=plt.gcf(); fig_state_piechart.set_size_inches(12,10);
```

```python
# introduce mutation
mutant_simulation = maboss.copy_and_mutate(model, ["K_N_HRAS"], "ON")
mutant_results = mutant_simulation.run()
mutant_results.plot_piechart()
fig_state_piechart=plt.gcf(); fig_state_piechart.set_size_inches(12,10);
```

```python
# nodes_probtraj=master_results.get_nodes_probtraj()
nodes_probtraj_df=pd.DataFrame(mutant_results.get_nodes_probtraj())

nodes_probtraj_df.loc[:,nodes_probtraj_df.columns.str.contains(
    '|'.join(['RAS','BRAF','CHEK1','MK2','mitos','apopt','senesc','DSB']),case=False)].plot(linewidth=3) 
# 'p38','ERK'
fig_nodetraj=plt.gcf(); fig_nodetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5),loc="center left",borderaxespad=0)
```
