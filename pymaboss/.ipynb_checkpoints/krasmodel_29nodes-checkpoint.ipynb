```python
import maboss
import ginsim
import pypint

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load model
# python bnet_to_maboss.py krasmodel_29nodes.bnet
# krasmodel_29nodes.bnd: 29 node model, where cdc25b had no upstream input
# charitemodel_v2.bnd: paths to cdc25b, erk1_2 -> mk2 link
# charitemodel_v2_cdk1_relaxed: CDK1 2 inputs enough
model = maboss.load("charitemodel_v2_cdk1_relaxed.bnd", "charitemodel_v2_cdk1_relaxed.cfg")
# model.get_logical_rules()
```

```python
# model.network
```

```python
# maboss.set_nodes_istate(model,['Senescence','DSB_SSB','Apoptosis_CC3'],{(0,0,0): 1})
# maboss.set_nodes_istate(model,['Senescence','mitotic_catastrophe','Mitosis','Apoptosis_CC3'], [1,0])

# set cell fate+other nodes to 0
nodes=('Senescence','mitotic_catastrophe','Mitosis','Apoptosis_CC3',
  'CDC25A','CDC25B','CDC25C','CDK1','CDK2','CDK4_6',
       'MK2_HSP27','K_N_HRAS','RPS6KA1_CSNK2A1','MEK','ERK1_2','p38',
       'ATR','ATM','TP53','CHEK1','CHEK2','HRR2','FAHRR')
istate = [1, 0]
for nd in nodes:
    model.network.set_istate(nd, istate)

# set input nodes to 1
model.network.set_istate('GFR', [0,1])
model.network.set_istate('DSB_SSB', [0,1])
```

```python
# model.network
# show initial values
# model.network.get_istate()
# model.network['MEK'].is_internal = True
maboss.set_output(model,['HRR2','Senescence','DSB_SSB','mitotic_catastrophe','Mitosis','mitotic_catastrophe',
    'MK2_HSP27','TP53','Apoptosis_CC3','ATM','ATR','CHEK1','p38','CDC25A','CDC25B','CDC25C','CDK1','CDK2','CDK4_6'])
model.update_parameters(discrete_time=0, use_physrandgen=0,
                            seed_pseudorandom=100, sample_count=500,
                            max_time=30, time_tick=0.1, thread_count=4,
                            statdist_traj_count=100, statdist_cluster_threshold=0.9)
master_results = model.run()
```

```python
nodes_probtraj=master_results.get_nodes_probtraj()
nodes_probtraj_df=pd.DataFrame(master_results.get_nodes_probtraj())
# fig, ax = plt.subplots()
# plt.plot(nodes_probtraj,linewidth=3,label=nodes_probtraj.columns.values) # 
# fig_nodetraj, ax = plt.subplots()
# ax = plt.subplot(111)
# nodes_probtraj_df.columns.str.contains('|'.join(['apopt','senesc','SSB','mit','CDK','CDC25']),case=False)
# fig, axs = plt.subplots(2)
nodes_probtraj_df.loc[:,nodes_probtraj_df.columns.str.contains(
    '|'.join(['CDK','CDC25','CHEK','MK2']),case=False)].plot(linewidth=3) 
# 'apopt','senesc','SSB','mit',
# axs[0].plot(nodes_probtraj_df.loc[:,nodes_probtraj_df.columns.str.contains('|'.join(['CDK','CDC25']),case=False)])
# nodes_probtraj_df.loc[:,'ATM'].plot(linewidth=3)
fig_nodetraj=plt.gcf(); fig_nodetraj.set_size_inches(10,6);
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
# plt.set_size_inches(12,7);
# nodes_probtraj.plt(kind='line')
# plt.legend();
```
